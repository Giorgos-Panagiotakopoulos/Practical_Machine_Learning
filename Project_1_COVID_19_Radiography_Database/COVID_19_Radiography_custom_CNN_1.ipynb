## Preproces Dataset Utils
Create a helper function in order to preproces datasets to fit in target model.
- Get image label from file path. **path eg:**  ( gs://kds-60ec5ed243a9d364cb261e8025dd35e48837e7ce35e3f0960c1c5b23/COVID-19_Radiography_Dataset/Lung_Opacity/Lung_Opacity-233.png ) we can see that the -2 element of the path is the image class.
- Load image as string raw data using *tf.io.read_file*
- Convert image from string data into 3D unit tensor
- conver image dType in to float in range [0, 1]
- resize image in target size


```python
# def get_img_lbl(path):

def transform_image(path):
  # get the label from file path
  label = 0 
  parts = tf.strings.split(path, os.path.sep)
  # The second to last is the class-directory
  # print(parts[-2] == "COVID")
  if(parts[-2] == "COVID"):
      label = 0
  elif(parts[-2] == 'Lung_Opacity'):
      label = 1
  elif(parts[-2] == 'Normal'):
      label = 2
  elif(parts[-2] == 'Viral Pneumonia'):
      label = 3
  else:
      label = 4

  # load and transform image
  img = tf.io.read_file(path)
  img = tf.image.decode_jpeg(img, channels=3)
  img = tf.image.convert_image_dtype(img, tf.float32)
  img = tf.image.resize(img, IMAGE_SIZE)
  return img, label
```

## Apply transformation on Train and test set
We apply transformation usig map() in order to apply transform_image helper fun on each file name. We also use and autotune prefetch buffer size for beater performance using TPU

```python
train_dataset = train_raw_dataset.map(transform_image, num_parallel_calls=AUTOTUNE)
val_dataset = val_raw_dataset.map(transform_image, num_parallel_calls=AUTOTUNE)
## we will use this iterator to prin dataset images (in batch of 9)
training_viz_iterator = train_dataset.batch(9).as_numpy_iterator()
```

## Prepare Datasets for Classification
- Cache train and test datasets (because we know that thay are small)
- Shaffle train dataset
- use repeat in order to Make sure that your dataset or generator can generate at least `steps_per_epoch * epochs` batches
- set batch size on both train and test datasets
- enable prefetch next batch while model model training with autotune prefetch buffer size

```python
train_dataset = train_dataset.cache().shuffle(2048).repeat().batch(BATCH_SIZE).prefetch(AUTOTUNE) 
val_dataset = val_dataset.cache().repeat().batch(BATCH_SIZE).prefetch(AUTOTUNE) 
```

# Create Custom CNN
- define model
- compile model
- fit datasets to model

```python
model = tf.keras.models.Sequential([
  BatchNormalization(),
  Conv2D(32, 3, activation='relu', input_shape=(299,299,3)),
  MaxPooling2D(),
  Conv2D(64, 3, activation='relu'),
  MaxPooling2D(),
  Dropout(0.3),
  Conv2D(128, 3, activation='relu'),
  MaxPooling2D(),
  Dropout(0.2),
  Flatten(),
  Dense(256, activation='relu'),
  Dropout(0.15),
  Dense(4, activation= 'softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy','mse'])
```

```python
history = model.fit(
    train_dataset,
    steps_per_epoch= TRAIN_COUNT // BATCH_SIZE,
    epochs=EPOCHS,
    validation_data= val_dataset,
    validation_steps= VAL_COUNT // BATCH_SIZE,
)

# model.save('customCNN.h5')
```

## Display Training results 


```python
def display_training_curves(training, validation, title, subplot):
  ax = plt.subplot(subplot)
  ax.plot(training)
  ax.plot(validation)
  ax.set_title('model '+ title)
  ax.set_ylabel(title)
  ax.set_xlabel('epoch')
  ax.legend(['training', 'validation'])

plt.subplots(figsize=(10,10))
plt.tight_layout()
display_training_curves(history.history['accuracy'], history.history['val_accuracy'], 'accuracy', 211)
display_training_curves(history.history['loss'], history.history['val_loss'], 'loss', 212)
```

## Create a DataFrame with the results


```python
# Get the model
with tpu_strategy.scope():

  models = {
      "MobileNetV2": {"model":tf.keras.applications.MobileNetV2(input_shape=(299, 299, 3), include_top= False, weights='imagenet'), "train_time":0},
      "DenseNet169": {"model":tf.keras.applications.DenseNet169(input_shape=(299, 299, 3), include_top= False, weights='imagenet'), "train_time":0},
      "DenseNet201": {"model":tf.keras.applications.DenseNet201(input_shape=(299, 299, 3), include_top= False, weights='imagenet'), "train_time":0},
      "InceptionV3": {"model":tf.keras.applications.InceptionV3(input_shape=(299, 299, 3), include_top= False, weights='imagenet'), "train_time":0},
      "InceptionResNetV2": {"model":tf.keras.applications.InceptionResNetV2(input_shape=(299, 299, 3), include_top= False, weights='imagenet'), "train_time":0},
      "ResNet50": {"model":tf.keras.applications.ResNet50(input_shape=(299, 299, 3), include_top= False, weights='imagenet'), "train_time":0},
      "VGG16": {"model":tf.keras.applications.VGG16(input_shape=(299, 299, 3), include_top= False, weights='imagenet'), "train_time":0},
      "VGG19": {"model":tf.keras.applications.VGG19(input_shape=(299, 299, 3), include_top= False, weights='imagenet'), "train_time":0}
  }

  local_epochs = EPOCHS 

  # Fit the models
  for name, model in models.items():
      
      m = model["model"]
      m.trainable = False
      configured_model = tf.keras.Sequential([
                                        m, 
                                        tf.keras.layers.Flatten(),
                                        tf.keras.layers.Dense(units=256, activation='relu'),
                                        tf.keras.layers.Dense(units=256, activation='relu'),
                                        tf.keras.layers.Dense(units=4, activation='softmax')
                                        ])
      
      configured_model.summary()

      configured_model.compile(
          optimizer='adam',
          loss='sparse_categorical_crossentropy',
          metrics=[
          'accuracy','mse'])

      # timeing trainging procidure
      start_training = time.time()

      history = configured_model.fit(
        train_dataset,
        steps_per_epoch=TRAIN_COUNT // BATCH_SIZE,
        epochs=local_epochs,
        validation_data=val_dataset,
        validation_steps=VAL_COUNT // BATCH_SIZE,
      )
      
      # save train duration to dict
      total_training = round((time.time() - start_training),2)
      models[name]['train_time'] = total_training

      # save val accuracy tou dict
      val_acc = history.history['val_accuracy']
      models[name]['val_accuracy'] = [round(v,4) for v in val_acc]
      models[name]['epoch_count'] = local_epochs

    # plot training results
      print("--------------------------------------------------------------------")
      print("|                                                                  |")
      print("|                                                                  |")
      print("{} trained in {} sec with val_acc: {}".format(name, total_training, models[name]['val_accuracy']))                                                              
      print("|                                                                  |")
      print("|                                                                  |")
      print("--------------------------------------------------------------------")

      plt.subplots(figsize=(10,10))
      plt.tight_layout()
      display_training_curves(history.history['accuracy'], history.history['val_accuracy'], 'accuracy', 211)
      display_training_curves(history.history['loss'], history.history['val_loss'], 'loss', 212)
      plt.show()

      # save model
      configured_model.save('{}/{}.h5'.format(TF_MODELS_PATH, name))
```

## Create a DataFrame with the results


```python
models_result = []

for name, v in models.items():
    models_result.append([ name, models[name]['val_accuracy'][-1], 
                          models[name]['train_time'],
                          models[name]['epoch_count']])
    
df_results = pd.DataFrame(models_result, 
                          columns = ['model','val_accuracy','Training time (sec)', 'Total Epoch'])
df_results.sort_values(by='val_accuracy', ascending=False, inplace=True)
df_results.reset_index(inplace=True,drop=True)
df_results
```
